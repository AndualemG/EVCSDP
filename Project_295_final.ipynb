```python
Data=pd.DataFrame()
Data['car_ID']=Car_Data.iloc[:,0]
Data['stop_lon']=Car_Data.iloc[:,1]
Data['stop_lat']=Car_Data.iloc[:,2] 
Data['start_stop_d']=Car_Data.iloc[:,3]
Data['start_stop_tod']=Car_Data.iloc[:,4]
Data['end_stop_d']=Car_Data.iloc[:,5] 
Data['end_stop_tod']=Car_Data.iloc[:,6]
Data['prev_trip_dist']=Car_Data.iloc[:,7] 
Data['prev_trip_t']=Car_Data.iloc[:,8]
Data['stop_t']=Car_Data.iloc[:,9]
Data['accumulated_dist']=Car_Data.iloc[:,10] 
Data['energy_depleted']=Car_Data.iloc[:,11]
Data['power_demand']=Car_Data.iloc[:,12] 
Data['weeks']=Car_Data.iloc[:,13]
Data['day_of_week']=Car_Data.iloc[:,14] 
Data['hour_of_day']=Car_Data.iloc[:,15]
```

```python
def Theta_cal(d):
    Tue=pd.DataFrame()
    Tue['w1']=ARX1.iloc[:,d]
    Tue['w2']=ARX2.iloc[:,d]
    Y=np.array(ARX3.iloc[:,d].values)
    Phi=np.array(Tue.values)
    Theta = np.dot(np.dot(np.linalg.inv(np.dot(Phi.T, Phi)), Phi.T), Y)
    return (Theta)
```

```python
Compare=(ARX1.iloc[:,1])
MAE_ARX = []
D=list(range(1,8))
MAE=0
for d in D:
    #df3 = Array_4D_stop_number[Array_4D_stop_number.DOW==f]
    DF=pd.DataFrame()
    DF['1']=np.array(ARX1.iloc[:,d].values)
    DF['2']=np.array(ARX2.iloc[:,d].values)
    Parx = []
    MAE_= 0
    parx = (np.dot(DF,Theta_cal(d)))
    Parx.append(parx)
    D=list(range(1,25))
    plt.figure(figsize = (10,6))
    plt.plot(D,Parx[0], label='ARX_Data')

    plt.plot(D,(ARX4.iloc[:,1]).values, label='Test_Data')
    
    plt.xlabel('Hour of Day')
    plt.ylabel('Energy depleted')
    plt.legend()
    plt.title('ARX model and Test_data (mean)')
    plt.show()
        #P = float(Array_4D_stop_number.loc[(Array_4D_stop_number['Hour'] == h) & (Array_4D_stop_number['Day'] == f), 'energy_depleted'])
    MAE_ += abs(Compare - parx)
    MAE_ARX.append(MAE_/7)
print('Average MAE')    
print(sum(MAE_ARX))
```
